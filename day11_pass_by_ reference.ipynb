```python
class Point:
    
    def __init__(self,x,y):
        self.x_cod = x
        self.y_cod = y
        
    def __str__(self):
        return '({},{})'.format(self.x_cod,self.y_cod)

#distance between two points
    def euclidean_distance(self,other):
        return ((self.x_cod-other.x_cod)**2+(self.y_cod-other.y_cod)**2)**0.5
        
# distance between origine and point
    def distance_origine(self):
        return self.euclidean_distance(Point(0,0)) #create object inside the class generally create outside of the class  


class Line:
    def __init__(self,A,B,C):
        self.A = A
        self.B = B
        self.C = C
        
    def __str__(self):
        return '{}X + {}Y + {} = 0'.format(self.A,self.B,self.C)

    def points_lies_on_line(line,point):
        if line.A*point.x_cod + line.B*point.y_cod + line.C == 0:
            return 'lies on the line'
        else:
            return 'does not lies'

    def shortest_distance(line,point):
        return abs(line.A*point.x_cod+line.B*point.y_cod + line.C)/(line.A**2+line.B**2)**0.5
        
```

```python
p1 = Point(1,1)
l1 = Line(1,1,-2)
p2 = Point(4,5)
```

```python
p1.euclidean_distance(p2)
```

```python
l1.points_lies_on_line(p1)
```

```python
l1.shortest_distance(p1)
```

## Reference Variable
    reference variable hold the objects
    we can create objects without reference variable as well
    an object can have multiple reference variable
    assigning a new referece variable to an existing object does not create a new object

```python
# object without a reference
class Person:
    def __init__(self):
        self.name = 'chhavi'
        self.gender = 'male'
        
# p is a reference variable, not an object , it contains the address of the object.
p = Person()
q = p
```

```python
print(p.name)
print(q.name)
q.name = 'money'
print(p.name)
print(q.name)
```
