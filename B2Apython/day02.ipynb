```python
# file I/O is done in most programming language ->open file ->read/write data-> close the file
#case 1 if the is not present
f = open("sample.txt","w")
f.write("hellow world")
f.close()
f.write("namastey")
```

```python
#write multiple strigns
f = open('sample.txt','w')
f.write('namstey')
f.write('\nnamstey maharaj')
f.close()
```

```python
#case2 if the file is already present
f = open('sample.txt','w')
f.write("simple is better than complex")
f.close()
```

```python
#how exactly works open()
f = open('sample.txt','a')
f.write('\n complex is better than complicate')
f.close()
```

```python
# write lines
L = ['himanchal pradesh \n','delhi\n','chhattishgrah']
f = open('sample.txt','w')
f.writelines(L)
f.close()
```
