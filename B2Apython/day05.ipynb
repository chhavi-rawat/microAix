### sets:
    unordered
    mutable
    no duplicates
    cannot contain mutable data type

```python
#create
#empty set
s = set()
print(type(s))
#1d and 2d set
s1 = {1,2,3}
print(s1)
s2 = {1,'hellow',4.4,True,(1,2,3)}
print(s2)
s3 = set([1,2,3])
print(type(s3))
#sets cannot have mutable items
s4 = {1,2,3,[4,5,6]}
print(s4
```

```python
s1 = {1,2,3}
s2 = {2,1,3}
print(s1==s2)
# accessing items: it doesn't work
s1[1]
```

```python
#editing items: it also doesn't work.
s1 = {1,2,3}
s1[1]=200
```

```python
#adding items
s = {1,2,3}
#add
s.add(4)
print(s)
#update: multiple items add in a single time
s.update([5,6,7])
print(s)
```

```python
#deleting items
s = {1,2,3,4}
print(s)
#discard: delete the item if there is no items what i want to delete then it doesn't throw error.
s.discard(1)
print(s)
#pop
s.pop()
print(s)
s.clear()
print(s)
```

sets operations

```python
s1 = {1,2,3,4}
s2 = {3,4,5,6}
#union(|)
print(s1|s2)
#intersection: common items
print(s1&s2)
#difference(-)
print(s1-s2) # items of s1 is not present in s2
print(s2-s1) # items of s2 is not present in s1
# symmetric differece: print all elements except common
print(s1^s2)
#memebership 
print(1 in s1)
#loops
for i in s1:
    print(i)
```

```python
#len/sum/min/max/sorted
s = {1,2,4,5,3,5,6,7,9}
print(len(s))
print(min(s))
print(max(s))
print(sorted(s))
```

```python
#union/update
s1 = {1,2,3,4}
s2 = {3,4,5,6}
# s1.union(s2)
s1.update(s2) # permanent change in s1
print(s1)
print(s2)
```

```python
#intersection/intersection_update
s11 = {1,2,3,4}
s22 = {3,4,5,6}
s11.intersection(s22)
print(s11)
s11.intersection_update(s22) #permanent changes in s1
print(s11)
print(s22)
```

```python
#symmetric_difference/symmetric_difference_update
s12 = {1,2,3,4}
s13 = {3,4,5,6}
s12.symmetric_difference(s13)
print(s12)
print(s13)
s12.symmetric_difference_update(s13) #s12 store in s13
print(s12)
print(s13)
```

```python
#isdisjoint/issubset/issuperset
s1 = {1,2,3,4}
s2 = {3,4,5,6}
s1.isdisjoint(s2) #disjoint: no common items
s1.issubset(s2)
s1.issuperset(s2)
```

```python
#copy
s1 = {1,2,3,4,5}
s2 = s1.copy()
print(s1)
print(s2)
```

```python
#frozen set: it is just an immutable version of a python set object. we can use where we need to read only operation.
#create frozen set
fs = frozenset([1,2,3,frozenset([1,2,3,4,5])])
fs1 = frozenset([4,5,6])
fs | fs1
```
