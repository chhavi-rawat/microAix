### dictionary: a collection of keys values, used to store data values like a map.
            dict = {'fruit':'grapes','vegtables':'chilly}
### CHRACTERISTICS:
            . mutable
            . no indexing work
            . keys cannot duplicate
            . keys cannot be mutable items

#### create dictionary

```python
#empty dictinary
d = {}
# 1D dictionary
d1 = {'name':'chhavi','title':'rawat','gender':'male'}
#with mixed keys
d2 = {(1,2,3):2,'radhe':'krishna'}
# 2d dictionary
d3 = {
    'name':'chhavi',
    'university':'CSVTU',
    'sem': 5,
    'subjects':{
        'dsa':56,
        'maths':44,
        'c++':57
    }
}
# using sequence and dict fuction
d4 = dict([('name','chhavi'),('age',24)])
# duplicate keys : no duplicate keys, first update whichever is last keys
d5 = {'name':'chhavi','name':'rawat','gender':'male'}
# mutable items as keys
d6 = {'name':'chhavi','title':'rawat','gender':'male'}
d6['name'] = 'rahul'
d6
```

### Accessing items

```python
birthday = {'day':12,'month':'july','year':2000}
print(birthday['month'])
#get 
birthday.get('year')
# how to extract marks of math subject.
marks_university = {
    'name':'chhavi',
    'university':'CSVTU',
    'sem': 5,
    'subjects':{
        'dsa':56,
        'maths':44,
        'c++':57
    }
}
marks_university['subjects']['maths']
```

### Adding keys-value pair

```python
birthday = {'day':12,'month':'july','year':2000}
birthday['university']='CSVTU'
birthday
```

### Remove key-values pair

```python
dob = {'day':12,'month':'july','year':2000}
# pop
dob.pop('month')
# popitem: delete last key-value 
dob.popitem()
dob
# clear
dob.clear()
# del
del dob
dob
```

### Dictionary Comprehension
    {key:value for var in iterable}

```python
#print 1 st to 10 and their square
{i:i**2 for i in range(1,11)}
```

```python
#using existing dictionary
marks = {'english':80,'maths':100}
{key:value * 0.34 for (key,value) in marks.items()}
```

```python
#using zip
days = ['sunday','monday','tuesday','thirsday','friday','saturday']
tem_c = [23.4,33,23,34,34,12,23.3]
{i:j for (i,j) in zip(days,tem_c)}
```

```python
#using if condition
products = {'phone':100,'charger':0,'tablet':0,'headphone':234}
{key:value for (key,value) in products.items() if value>0}
```

```python
#nested comprehension
# print tables of number form 2 to 4
{i:{j:i*j for j in range(1,11)} for i in range(2,11)}
```
