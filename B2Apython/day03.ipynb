```python
#creating list
#empty
print([])
#1D list -homogeneous
print([1,2,3,4])
#2D list
print([[1,2,3],[4,5,6]])
#3D list
print([[[1,2],[1,2],[1,2,3]]])
print(list('hellow'))
```

```python
#accessing items from a list
#indexing(positive, negative indexing)
l = [1,2,3,5]
# positive indexing: it goes from left to write (0,1,2,3...n)
print(l[0])
print(l[3])
#negative indexing: it goes from right to left of the list(-1,-2,-3...........n)
print(l[-1])
#slicing
print(l[0:4])
print(l[::-1]) #reverse 
```

#### adding item to list

```python
# append
l = [1,2,3]
l.append(4)
print(l)
#insert
l.insert(0,99)
print(l)
l.extend([4,5,6])
print(l)

```

#### editing items 

```python
l = [1,2,3,4]
# indexing
l[0]=10
print(l)
l[1:4]=[20,30,40]
print(l)
```

#### deleting items

```python
#remove
l = [10,20,30]
print(l)

#remove
l.remove(10)
print(l)

#pop
l.pop(0)
print(l)

#clear
l.clear()
print(l)

#del
del l
print(l)
```

#### operations on lists

```python
#arithmetic
l1 = [1,2,3]
l2 = ['choice']
print(l1+l2)
print(4*l2)

#membership 
print(2 in l1)
print(2 not in l1)
```

#### list functions

```python
#count/len/min/max/sorted
l = [1,2,3,4,5,6,1,1]
# count: count frequncey
print(l.count(1))
print(len(l))
print(l.index(1))
print((1))
#reverse: permanently changes
l1=l.reverse()
print(l)
#sorted: ascending order
print(sorted(l))
# sort: permanently sorted
l.sort()
print(l)

#copy: create copy in memory
country = ['japan','australia','india','nepal','shrilanka']
c1 = country.copy()
print(c1)
c1.reverse()
print(c1)
c1.sort()
print(c1)
print(country)
```
