## variables: 
    variable is a container that stores a value, somewhat like a bucket.in python
     variables aren't buckets.instead, they are lables or tags that refer to objects in the python interpreter's namespace.

```python
var = "variables"
var1 = 12
var2 = 3.4
var3 = True
print(var,var1,var2,var3)
```

```python
#operators are used to perform operations on variables and values

#arighmetic operators: addition(+), subtraction(-), multiplication(*), division(/)
                        # modulus(%), exponentiation(**), floor division(//)

a = 5
b = 2
#addition
print(a+b)

#subtraction
print(a-b)

#multiplication
print(a*b)

#division
print(a/b)

#exponentiation
print(a**b)

#floor divion: return neareat whole number
print(a//b)
```

```python
#assignment operators: used to assign values to variables.
                        # (=, +=, -+, *=, /=, %=, //=, **=, &=, |=, ^=,>>=,<<=, :=)
x = 2
print(x)
# x = x+3
x+=3
print(x)

# creating y varible
y = 21

# y = y-1
y-=1
print(y)
```
