```python
def is_even(num):
    ''' 
    this fuction returns if a given num is odd or even
    input - any valid integer
    ouput - odd/even
    create by - chhavi rawat
    date 28/11/25
    '''
    if num % 2 == 0:
        return 'even'
    else:
        return 'odd'
```

#### *args and **kwargs
    they are special python keywords that are used to pass the variable lenth of arguments to a function
    

```python
#args: allows to pass a variable number of non-keyword arguments to a function
def addition(*god):
    add = 0

    for i in god:
        add = add+i

    return add
```

```python
addition(1,2,3,4,5,6,7,8)
```

#### Lambda function
    a small anonymous function, it can take any number of arguments, but can only have one expression.
##### difference between lambda vs Normal function
    . no name
    . lambda has no return value
    . lambda is written in one line
    . not reusable
##### they are used with HOF: a function who return a fuction. one function return to another function.
    

```python
# check if a stirng has 'a'
a = lambda s:'a' in s
a('hellow')
```

```python
#odd/even
a = lambda x :'even'if x%2==0 else 'odd'
a(5)
```

#### Higher Order Function

```python
def square(x):
    return x**2
    
# HOf    
def transform(f,L):
    output = []
    for i in L:
        output.append(f(i))
    print(output)
L = [1,2,3,4]
# transform(square,L)
transform(lambda x: x**3,L)
```

```python
#map: lambda/iterable
list(map(lambda x: x**3,[1,2,3,4]))
```

```python
#square labelling of list items
L = [3,4,5]
list(map(lambda x: 'even' if x%2==0 else 'odd',L))
```

```python
users = [
    {
        
        'name':'chhavi',
        'state':'chhattishgrah'
    },
    {
        
        'name':'ashu',
         'state':'up'
    },
    {
        'name':'aruna',
        'state':'himanchal pradesh'
    }
]

# extract name of the users
list(map(lambda users: users['state'],users))
```

### Fillter: filltering from a list

```python
L = [3,4,55,6,7,889,9,99,65]
list(filter(lambda x: x>6,L))

fruits = ['apple','banana','orange','grapes']
list(filter(lambda x :x.startswith('a'),fruits))
```
