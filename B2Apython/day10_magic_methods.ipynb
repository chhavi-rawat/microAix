```python
read = "read the manula to use"
print(f"{read.upper()}","\n======================",Fraction.__doc__)
class Fraction:
    # docstrings
    """     
                    Fraction operation: add,subract,mulfiply,divide and simplify each of then
                    Numerator and Denomenator: should be : non_negative number
                    Create Fraction: f1 = Fraction(12,5)
                                     f2 = Fraction(6,9)
                                eg.  print(f1*f2)
                """
    def __init__(self,x,y):
        self.num = x
        self.den = y
        
    
    def __str__(self):
        return f'Fraction:{(self.num)}/{(self.den)}'
        
    def fraction_to_decimal(self):
        return '{}'.format(self.num/self.den)
        
    def __add__(self,other):
        return f'Addition of Fraction: {self.num*other.den+self.den*other.num}/{self.den*other.den}'
    def simplyfy_add(self,other):
        return f'Addition of Fraction: {(self.num*other.den + self.den*other.num)/(self.den*other.den)}'

    def __sub__(self,other):
        return f'sub of Fraction: {self.num*other.den-self.den*other.num}/{self.den*other.den}'
    def simplyfy_subraction(self,other):
        return f'subtraction of Fraction: {(self.num*other.den-self.den*other.num)/(self.den*other.den)}'
        
    def __mul__(self,other):
        return f'sub of Fraction: {self.num*other.num}/{self.den*other.den}'
    def simpyfy_multiplication(self,other):
        return f'sub of Fraction: {(self.num*other.num)/(self.den*other.den)}'
        
    def __truediv__(self,other):
        new_num = self.num*other.den
        new_den = self.den*other.num
        return f'{new_num}/{new_den}'
        
    def simplyfy_div(self,other):
        return (self.num*other.den)/(self.den*other.num)          
```

```python
f1 = Fraction(3,4)
f2 = Fraction(4,5)
```

```python
print(f1+f2)
f1.simplyfy_add(f2)
```

```python
print(f1-f2)
f1.simplyfy_subraction(f2)
```

```python
print(f1*f2)
f1.simpyfy_multiplication(f2)
```

```python
print(f1/f2)
f1.simplyfy_div(f2)
```
