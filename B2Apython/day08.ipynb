### class: it is a blueprint of object.
### object: it is a instance of class.

```python
class Atm:
    #docstrings
    '''    
    Atm class simulates basic ATM operations.
    
    Methods:
        create_pin()   - Allows user to set a PIN and initial balance.
        change_pin()   - User can change the pin.
        check_balance()- Displays current balance after PIN verification.
        withdraw_balance()- Withdraws money if sufficient funds are available.
    '''    
    # constructor
    def __init__(self):
        self.pin = ''
        self.balance = 0
        self.menu()
        
    def menu(self):
        user_input = input('''
        how can i help you
        1. press 1 to create pin
        2. press 2 to change pin
        3. press 3 to check balance
        4. press 4 to withdraw 
        5. anything else to exit
        ''')
        if user_input == '1':
            #create pin
            self.create_pin()
        elif user_input == '2':
            #change pin
            self.change_pin()
        elif user_input == '3':
            #check balance
            self.check_balance()
        elif user_input == '4':
            #withdraw blance
            self.withdraw_balance()
           
        else:
            exit()
            
     
    def create_pin(self):
        user_pin = input('create your pin')
        self.pin = user_pin
        user_balance = int(input('enter your balance'))
        self.balance = user_balance
        print('pin create successfully')
        self.menu()

    def change_pin(self):
        old_pin = input('enter your existing pin')
        if self.pin == old_pin:
            new_pin = input('create new pin')
            self.pin = new_pin
            print('pin change succsseful')
            self.menu()
        else:
            print('galat pin dala gya')
            self.menu()
            
    def check_balance(self):
        user_pin = input('enter your pin')
        if self.pin == user_pin:
            print('your balance is:',self.balance)
        else:
            print('incorrect pin')
        self.menu()
        
    def withdraw_balance(self):
        user_pin = input('enter your existing pin')
        if user_pin == self.pin:
            amount = int(input('enter the amount'))
            if amount <= self.balance:
                self.balance = self.balance - amount
                print('withdrawl successfull , now your balance is:',self.balance)
                self.menu()
            else:
                print('Not sufficient fund')
                self.menu()
        else:
            print('your pin is incorrect')
        self.menu()                   
```
