## Encapsulation

```python
#Encapsulation
class Atm:
    def __init__(self):
        self.__pin = ''
        self.__balance = 0

    def getBalance(self):
        return self.__balance

    def setBalance(self,newValue):
        if type(newValue) == int:
            self.__balance = newValue
        else:
            print('beta bhut marenge')

    def __menu(self):
        user_input = input('''
        how can i help you ?
        1. Press 1 to create pin.
        2. Press 2 to change pin.
        3. Press 3 to check balancee.
        4. Press 4 to withdraw balancee.
        5. Press 5 to exit.
        ''')
        if user_input == '1':
            self.createPin()
        elif user_input == '2':
            self.changePin()
        elif user_input == '3':
            self.checkBalance()
        elif user_input == '4':
            self.withdrawBalance()
        else:
            exit()


    def createPin(self):
        userPin = input('Create pin:')
        userBalance = int(input('Enter balancee'))
        self.__pin = userPin
        self.__balance = userBalance
        print('Pin create successful')

    def changePin(self):
        userPin = input('Enter existing pin')
        if userPin == self.__pin:
            newPin = input('Enter new pin')
            self.__pin = newPin
            print('pin changed successful')
        else:
            print('Galat pin dala gaya')

    def checkBalance(self):
        userPin = input('Enter existing pin')
        if userPin == self.__pin:
            print('balance is:',self.__balance)
        else:
            print('Galat pin data gaya')

    def withdrawBalance(self):
        userPin = input('Enter your pin')
        if userPin == self.__pin:
            userBalance = int(input('How much balancee want to withdraw'))
            if self.__balance>= userBalance:
                self.__balance = self.__balance - userBalance
                print(f'''
                            Apke Atm me total Rs. {self.__balance}
                            Rs.{userBalance} nikala hain
                            Atm me : RS.{self.__balance} . Dhan hain''')
            else:
                print('Apke pass prayapt Dhan nhin hai')
        else:
            print('Galat Pin Dala gya')
       
```

```python
#static variable 
class Person:
    __count = 1

    def __init__(self,name,age):
        self.name = name
        self.age = age
        self.__pid = Person.__count
        Person.__count +=1
        
    def getPid(self):
        return self.__pid
        
p1 = Person('avni',23)
p2 = Person('anu',22)
p3 = Person('maya',24)
```
